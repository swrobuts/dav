## 🗄️ Datenbankverbindung testen

````python
import sqlite3
import os

# SQLite Test
db_path = '../data/dav_sample.db'
if os.path.exists(db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
    tables = cursor.fetchall()
    print('✅ SQLite verbunden!')
    print('   Tabellen:', [t[0] for t in tables])
    conn.close()
else:
    print('⚠️ dav_sample.db nicht gefunden.')
    print('   Führe aus: cd /home/jovyan/data && python generate_data.py')
````

## 📁 Datei-Übersicht

````python
import os
import pandas as pd

data_dir = '../data'
print('=== Verfügbare Datendateien ===')
print()

files = []
for f in sorted(os.listdir(data_dir)):
    path = os.path.join(data_dir, f)
    if os.path.isfile(path):
        size_kb = os.path.getsize(path) / 1024
        files.append({'Datei': f, 'Größe (KB)': f'{size_kb:.1f}'})

df_files = pd.DataFrame(files)
print(df_files.to_string(index=False))
````

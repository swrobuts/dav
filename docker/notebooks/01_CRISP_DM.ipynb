## 🔄 CRISP-DM: Cross-Industry Standard Process for Data Mining

CRISP-DM beschreibt den Standard-Prozess für Data-Science-Projekte. Es ist kein linearer Prozess – die Phasen überlappen und man springt zwischen ihnen.

```
┌─────────────────────────────────────────────────────────────┐
│                        CRISP-DM                             │
│                                                             │
│   1. Business          ←──────────────────→   6. Deployment │
│      Understanding                                          │
│         │                                         ↑         │
│         ↓                                         │         │
│   2. Data              ←──────────────────   5. Evaluation  │
│      Understanding                                          │
│         │                                         ↑         │
│         ↓                                         │         │
│   3. Data Preparation  ────────────────→   4. Modeling     │
│      *** DAV! ***                                           │
└─────────────────────────────────────────────────────────────┘
```

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# CRISP-DM Phasen visualisieren
phases = [
    ('Business\nUnderstanding', '#0389CF', 0.35),
    ('Data\nUnderstanding', '#06b6d4', 0.30),
    ('Data\nPreparation\n★ DAV ★', '#E8792F', 0.20),
    ('Modeling', '#8b5cf6', 0.08),
    ('Evaluation', '#22c55e', 0.05),
    ('Deployment', '#f59e0b', 0.02),
]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Kreisdiagramm: Zeitanteil pro Phase
labels = [p[0] for p in phases]
sizes = [p[2] for p in phases]
colors = [p[1] for p in phases]
explode = [0.1 if 'DAV' in l else 0 for l in labels]

ax1.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.0f%%', startangle=90, textprops={'fontsize': 9})
ax1.set_title('Typischer Zeitanteil pro CRISP-DM-Phase\n(Real-World-Projekte)', 
              fontsize=12, fontweight='bold')

# Balkendiagramm: Phase-Beschreibungen
phase_names = [p[0].replace('\n', ' ').replace('★ DAV ★', '(DAV)') for p in phases]
y_pos = range(len(phases))
bars = ax2.barh(y_pos, [p[2]*100 for p in phases], color=colors, edgecolor='white', height=0.6)
ax2.set_yticks(y_pos)
ax2.set_yticklabels(phase_names, fontsize=10)
ax2.set_xlabel('Zeitanteil (%)', fontsize=11)
ax2.set_title('CRISP-DM: Zeitaufwand je Phase', fontsize=12, fontweight='bold')
ax2.invert_yaxis()

for bar, size in zip(bars, sizes):
    ax2.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2,
             f'{size*100:.0f}%', va='center', fontsize=9)

plt.tight_layout()
plt.savefig('../data/crisp_dm_chart.png', dpi=100, bbox_inches='tight')
plt.show()
print('💡 Data Preparation (DAV) macht ~20% aus – aber schlechte Vorbereitung ruiniert die anderen 80%!')
````

## 🧭 Wo steht DAV im CRISP-DM?

| Phase | Aktivitäten | DAV-Bezug |
|-------|------------|----------|
| Business Understanding | Ziele definieren, KPIs festlegen | Kontext verstehen |
| **Data Understanding** | **EDA, erste Analysen, Qualitätsprüfung** | **NB 04** |
| **Data Preparation** | **Bereinigen, Transformieren, Integrieren** | **NB 05, 06, 07** |
| Modeling | ML-Modelle trainieren | Feature Engineering (NB 06) |
| Evaluation | Modell bewerten | Outputs interpretieren |
| Deployment | Produktivbetrieb | – |

````python
# Mini-Demo: CRISP-DM in Python – ein einfaches Beispiel
import pandas as pd
import numpy as np

print('=== CRISP-DM Demo: Flugverspätungen ===')
print()

# 1. Business Understanding
print('📌 Phase 1 – Business Understanding')
print('   Frage: Welche Flüge haben die höchsten Verspätungen?')
print('   KPI: Durchschnittliche Verspätung pro Airline')
print()

# 2. Data Understanding
print('📌 Phase 2 – Data Understanding')
df = pd.read_csv('../data/flights.csv')
print(f'   Datensatz: {df.shape[0]:,} Flüge, {df.shape[1]} Spalten')
print(f'   Fehlende Werte: {df.isnull().sum().sum():,}')
print(f'   Airlines: {df["carrier"].nunique()}')
print()

# 3. Data Preparation
print('📌 Phase 3 – Data Preparation (DAV)')
df_clean = df.dropna(subset=['dep_delay', 'arr_delay', 'carrier'])
print(f'   Nach Bereinigung: {len(df_clean):,} Zeilen ({len(df)-len(df_clean)} entfernt)')
print()

# 4. (einfaches) Modeling/Analysis
print('📌 Phase 4 – Analyse/Modeling')
avg_delay = df_clean.groupby('carrier')['dep_delay'].agg(['mean','count']).round(1)
avg_delay.columns = ['Ø Verspätung (min)', 'Anzahl Flüge']
avg_delay = avg_delay.sort_values('Ø Verspätung (min)', ascending=False)
print(avg_delay.head(10).to_string())
````

## 💡 Wichtige Erkenntnisse

- **CRISP-DM ist iterativ**: Du wirst zwischen Phasen springen
- **Data Preparation ist der Schlüssel**: 80% der Zeit, aber oft unterschätzt
- **"Garbage in, Garbage out"**: Schlechte Daten → schlechte Ergebnisse
- **DAV deckt Phasen 2+3 ab**: Data Understanding und Data Preparation

````python
# ✅ Challenge: Interaktive Frage
import ipywidgets as widgets
from IPython.display import display

print('=== Challenge: CRISP-DM ===')
print()
print('Frage: Du entdeckst beim Modellieren, dass deine Verspätungsdaten')
print('       viele fehlende Werte haben. Welche Phase rufst du auf?')
print()

options = [
    'Direkt zur Deployment-Phase gehen',
    'Zurück zu Data Preparation (Data Cleansing)',
    'Das Modell einfach mit NaN-Werten trainieren',
    'Das Projekt abbrechen'
]

radio = widgets.RadioButtons(options=options, description='')
button = widgets.Button(description='Antwort prüfen', button_style='primary')
output = widgets.Output()

def check(btn):
    with output:
        output.clear_output()
        if radio.value == options[1]:
            print('✅ Richtig! CRISP-DM ist iterativ. Du gehst zurück zu Data Preparation.')
        else:
            print(f'❌ Falsch. Richtig wäre: "{options[1]}"')

button.on_click(check)
display(radio, button, output)
````

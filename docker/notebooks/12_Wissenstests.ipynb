## 🎯 Teil 1: Grundlagen (8 Fragen)

Führe die folgende Zelle aus, um den interaktiven Quiz zu starten:

```python
def create_quiz_section(questions, section_title):
    """Erstellt einen interaktiven Quiz-Abschnitt."""
    section_q = [q for q in questions if q['section'] == section_title]
    
    print(f'\n=== {section_title} ({len(section_q)} Fragen) ===')
    
    radio_widgets = []
    for i, q in enumerate(section_q):
        radio = widgets.RadioButtons(
            options=q['opts'],
            description='',
            layout=widgets.Layout(width='100%')
        )
        label = widgets.HTML(f'<b>Frage {i+1}/{len(section_q)}:</b> {q["q"]}')
        display(label, radio)
        radio_widgets.append((radio, q['ans'], q['opts']))
    
    check_btn = widgets.Button(
        description=f'✓ {section_title} auswerten',
        button_style='primary',
        layout=widgets.Layout(width='300px', margin='10px 0')
    )
    result_out = widgets.Output()
    
    def on_check(btn):
        with result_out:
            clear_output()
            score = 0
            for j, (radio, correct_idx, opts) in enumerate(radio_widgets):
                selected = radio.value
                correct = opts[correct_idx]
                if selected == correct:
                    score += 1
                    print(f'  ✅ Frage {j+1}: Richtig!')
                else:
                    print(f'  ❌ Frage {j+1}: Falsch. Richtig: {correct}')
            total = len(radio_widgets)
            pct = score / total * 100
            print(f'\n📊 Ergebnis: {score}/{total} ({pct:.0f}%)')
            if pct >= 80:
                print('🏆 Ausgezeichnet! Weiter so!')
            elif pct >= 60:
                print('👍 Gut! Noch etwas Übung nötig.')
            else:
                print('📖 Bitte noch einmal die Theorie lesen.')
    
    check_btn.on_click(on_check)
    display(check_btn, result_out)

# Teil 1 starten
create_quiz_section(ALL_QUESTIONS, 'Teil 1: Grundlagen')
```

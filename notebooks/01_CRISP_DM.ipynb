## 🔄 CRISP-DM: Cross-Industry Standard Process for Data Mining

CRISP-DM beschreibt den Standard-Prozess für Data-Science-Projekte. Es ist kein linearer Prozess – die Phasen überlappen und man springt zwischen ihnen.

```
┌─────────────────────────────────────────────────────────────┐
│                        CRISP-DM                             │
│                                                             │
│   1. Business          ←──────────────────→   6. Deployment │
│      Understanding                                          │
│         │                                         ↑         │
│         ↓                                         │         │
│   2. Data              ←──────────────────   5. Evaluation  │
│      Understanding                                          │
│         │                                         ↑         │
│         ↓                                         │         │
│   3. Data Preparation  ────────────────→   4. Modeling     │
│      *** DAV! ***                                           │
└─────────────────────────────────────────────────────────────┘
```

````python
BASE_URL = "https://raw.githubusercontent.com/swrobuts/dav/main/data/"
````

## 🧭 Wo steht DAV im CRISP-DM?

| Phase | Aktivitäten | DAV-Bezug |
|-------|------------|----------|
| Business Understanding | Ziele definieren, KPIs festlegen | Kontext verstehen |
| **Data Understanding** | **EDA, erste Analysen, Qualitätsprüfung** | **NB 04** |
| **Data Preparation** | **Bereinigen, Transformieren, Integrieren** | **NB 05, 06, 07** |
| Modeling | ML-Modelle trainieren | Feature Engineering (NB 06) |
| Evaluation | Modell bewerten | Outputs interpretieren |
| Deployment | Produktivbetrieb | – |

````python
# Mini-Demo: CRISP-DM in Python – ein einfaches Beispiel
import pandas as pd
import numpy as np

print('=== CRISP-DM Demo: Flugdaten ===')
print()

# 1. Business Understanding
print('📌 Phase 1 – Business Understanding')
print('   Frage: In welchen Monaten gibt es das höchste Passagieraufkommen?')
print('   KPI: Durchschnittliche Passagiere pro Monat')
print()

# 2. Data Understanding
print('📌 Phase 2 – Data Understanding')
# Hinweis: BASE_URL + "flights.csv" scheint hier Zeitreihendaten zu liefern
df = pd.read_csv(BASE_URL + "flights.csv")
print(f'   Datensatz: {df.shape[0]:,} Zeilen, {df.shape[1]} Spalten')
print(f'   Spalten im Datensatz: {df.columns.tolist()}')
print(f'   Fehlende Werte: {df.isnull().sum().sum():,}')
print()

# 3. Data Preparation
print('📌 Phase 3 – Data Preparation (DAV)')
# Bereinigung: Wir stellen sicher, dass Passagierdaten vorhanden sind
df_clean = df.dropna(subset=['passengers', 'month'])
print(f'   Nach Bereinigung: {len(df_clean):,} Zeilen')
print()

# 4. (einfaches) Modeling/Analysis
print('📌 Phase 4 – Analyse/Modeling')
# Da 'carrier' fehlt, analysieren wir nach 'month'
avg_passengers = df_clean.groupby('month')['passengers'].agg(['mean','count']).round(1)
avg_passengers.columns = ['Ø Passagiere', 'Anzahl Datensätze']
avg_passengers = avg_passengers.sort_values('Ø Passagiere', ascending=False)
print(avg_passengers.head(12).to_string())
````

## 💡 Wichtige Erkenntnisse

- **CRISP-DM ist iterativ**: Du wirst zwischen Phasen springen
- **Data Preparation ist der Schlüssel**: 80% der Zeit, aber oft unterschätzt
- **"Garbage in, Garbage out"**: Schlechte Daten → schlechte Ergebnisse
- **DAV deckt Phasen 2+3 ab**: Data Understanding und Data Preparation

````python
# ✅ Challenge: Interaktive Frage
import ipywidgets as widgets
from IPython.display import display

print('=== Challenge: CRISP-DM ===')
print()
print('Frage: Du entdeckst beim Modellieren, dass deine Verspätungsdaten')
print('       viele fehlende Werte haben. Welche Phase rufst du auf?')
print()

options = [
    'Direkt zur Deployment-Phase gehen',
    'Zurück zu Data Preparation (Data Cleansing)',
    'Das Modell einfach mit NaN-Werten trainieren',
    'Das Projekt abbrechen'
]

radio = widgets.RadioButtons(options=options, description='')
button = widgets.Button(description='Antwort prüfen', button_style='primary')
output = widgets.Output()

def check(btn):
    with output:
        output.clear_output()
        if radio.value == options[1]:
            print('✅ Richtig! CRISP-DM ist iterativ. Du gehst zurück zu Data Preparation.')
        else:
            print(f'❌ Falsch. Richtig wäre: "{options[1]}"')

button.on_click(check)
display(radio, button, output)
````

```python
BASE_URL = "https://raw.githubusercontent.com/swrobuts/dav/main/data/"
```

```python
import pandas as pd
import numpy as np
import os

# Einfaches CSV
df_flights = pd.read_csv(BASE_URL + "flights.csv")
print(f'flights.csv: {df_flights.shape}')
print(df_flights.head(3))
print()

# CSV mit Semikolon-Trennzeichen
df_muster = pd.read_csv(BASE_URL + "muster_csv.csv", sep=';', encoding='utf-8', on_bad_lines='skip')
print(f'muster_csv.csv: {df_muster.shape}')
print(df_muster.head(3))
```

## 2️⃣ JSON – JavaScript Object Notation

```python
import json, urllib.request

# Flaches JSON
df_json = pd.read_json(BASE_URL + "muster_json.json")
print('muster_json.json:')
print(df_json.head(3))

# Verschachteltes JSON – temporäre Datei herunterladen
urllib.request.urlretrieve(BASE_URL + "complex_muster_json.json", "/tmp/complex_muster_json.json")
with open('/tmp/complex_muster_json.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
print()
print('complex_muster_json.json (Typ):', type(data))
if isinstance(data, list):
    print('Ersten Datensatz:', data[0] if data else '(leer)')
elif isinstance(data, dict):
    print('Keys:', list(data.keys())[:5])
```

## 3️⃣ XML – Extensible Markup Language

```python
import xml.etree.ElementTree as ET, urllib.request

urllib.request.urlretrieve(BASE_URL + "dataset.xml", "/tmp/dataset.xml")
tree = ET.parse('/tmp/dataset.xml')
root = tree.getroot()
print(f'XML Root-Tag: {root.tag}')
print(f'Anzahl Kinder: {len(root)}')
print()

# Alle Elemente extrahieren
records = []
for child in root:
    record = {elem.tag: elem.text for elem in child}
    records.append(record)

if records:
    df_xml = pd.DataFrame(records)
    print(f'DataFrame: {df_xml.shape}')
    print(df_xml.head(3))
else:
    for i, child in enumerate(root[:3]):
        print(f'Element {i}: tag={child.tag}, attribs={child.attrib}')
```

## 4️⃣ Excel – .xlsx Tabellen

```python
# Excel einlesen
df_excel = pd.read_excel(BASE_URL + "muster_xlsx.xlsx")
print(f'muster_xlsx.xlsx: {df_excel.shape}')
print(df_excel.head(3))

# Alle Sheets lesen
xl = pd.ExcelFile(BASE_URL + "muster_xlsx.xlsx")
print(f'\nSheets: {xl.sheet_names}')

# Verkaufsdaten mit Problemen – für Notebook 05!
df_prob = pd.read_excel(BASE_URL + "verkaufsdaten_mit_problemen.xlsx")
print(f'\nverkaufsdaten_mit_problemen.xlsx: {df_prob.shape}')
print('Fehlende Werte:')
print(df_prob.isnull().sum()[df_prob.isnull().sum()>0])
```

## 5️⃣ Parquet – Spaltenbasiertes Format

```python
# CSV als Parquet speichern
df_flights.to_parquet('/tmp/flights.parquet', index=False)

# Parquet lesen
df_parquet = pd.read_parquet('/tmp/flights.parquet')

# Größenvergleich
import io, requests, time
csv_resp = requests.get(BASE_URL + 'flights.csv')
csv_size = len(csv_resp.content) / 1024
import os
pq_size = os.path.getsize('/tmp/flights.parquet') / 1024
print(f'CSV:     {csv_size:.1f} KB')
print(f'Parquet: {pq_size:.1f} KB')
print(f'Parquet ist {csv_size/pq_size:.1f}x kleiner!')

# Geschwindigkeitsvergleich
import time
t1 = time.time(); pd.read_csv(BASE_URL + 'flights.csv'); t_csv = time.time()-t1
t1 = time.time(); pd.read_parquet('/tmp/flights.parquet'); t_pq = time.time()-t1
print(f'\nLese-Zeit CSV:     {t_csv*1000:.1f} ms')
print(f'Lese-Zeit Parquet: {t_pq*1000:.1f} ms')
```

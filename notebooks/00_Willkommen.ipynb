## 🗄️ Datenbankverbindung testen

````python
import sqlite3
import urllib.request
# Datenbankdatei von GitHub herunterladen
urllib.request.urlretrieve(BASE_URL + "dav_sample.db", "/tmp/dav_sample.db")
conn = sqlite3.connect("/tmp/dav_sample.db")
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cursor.fetchall()
print('✅ SQLite verbunden!')
print('   Tabellen:', [t[0] for t in tables])
conn.close()
````
